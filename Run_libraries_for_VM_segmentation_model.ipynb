```python
%%bash

wget -qO micromamba.tar.bz2 https://micro.mamba.pm/api/micromamba/linux-64/latest
tar -xjf micromamba.tar.bz2
mv bin/micromamba /usr/local/bin/
rm -rf bin micromamba.tar.bz2
micromamba --help | head -n 2
```

```python
%%bash

micromamba create -y -p /usr/local/envs/py311 python=3.11
```

```python
%%bash

micromamba run -p /usr/local/envs/py311 \
  pip install --index-url https://download.pytorch.org/whl/cu121 \
  torch==2.1.0+cu121 torchvision==0.16.0+cu121
```

```python
%%bash

micromamba run -p /usr/local/envs/py311 \
  pip install "mmengine==0.10.5" "mmdet==3.3.0"

micromamba run -p /usr/local/envs/py311 \
  pip install -f https://download.openmmlab.com/mmcv/dist/cu121/torch2.1.0/index.html "mmcv==2.1.0"
```

```python
%%bash


micromamba run -p /usr/local/envs/py311 python - << 'PY'
import sys, torch, mmcv, mmdet, mmengine, mmcv.ops
print("Python:", sys.version.split()[0])
print("torch:", torch.__version__)
print("mmcv:", mmcv.__version__, "| mmengine:", mmengine.__version__, "| mmdet:", mmdet.__version__)
print("CUDA avail:", torch.cuda.is_available(), "| DCN ops:", hasattr(mmcv.ops, "DeformConv2d"))
PY
```
